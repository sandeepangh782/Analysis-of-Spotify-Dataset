```python
#top 5 artist based on number of songs produced
top_five_artists=df.groupby("artist").count().sort_values(by="song name",ascending=False)['song name'][:5]
top_five_artists
```

```python
#pie chart representation of genre
df_genre = df['top genre'].value_counts()/len(df)
label=df_genre.index.tolist()
size=df_genre.values.tolist()
fig1, ax1 = plt.subplots(figsize=(10,10))
ax1.pie(size, labels=label, autopct='%1.1f%%', shadow=False, textprops={'fontsize': 8})
ax1.axis('equal')
plt.show()
```

```python
#bar graph representation of number of songs produced from year 2013 to 2017
year_song=df.groupby('year').count().sort_values(by='year',ascending=True)['song name']
fig=year_song.plot.barh()
fig.set_xlabel('Frequency of songs released')
fig.set_ylabel('Year')
plt.show()
```

```python
top_loudness=df[['dB','song name']].sort_values(by='dB',ascending=False)
top_loudness.set_index('song name')
```

```python
#plotting graph to between frequency of songs and loudness
fi=df.groupby('dB').count().sort_values(by='dB',ascending=False)['song name']
figure=fi.plot(kind='bar')
plt.show()
```

```python
#graphical representation of top 20 loudest tracks with their track names 
x=top_loudness['song name'][:20].tolist()
y=top_loudness['dB'][:20].tolist()
ax=sns.barplot(x=x,y=y)
ax.set_ylabel('Loudness')
ax.set_xticklabels(x,rotation=90)

plt.show()
```

```python
#histogram plot for duration of songs with respect to number of songs
fig = df['dur'].plot(kind='hist', bins=30)
fig.set_xlabel('Duration')
fig.set_ylabel('Number of songs')
plt.show()
```

```python
#scatter plot between energy and bpm among different genres
fig=sns.scatterplot(x='bpm',y='nrgy',data=df,hue='top genre',s=100)
sns.move_legend(fig, "upper left", bbox_to_anchor=(1, 1))
plt.show()
```

```python
fig5=sns.pairplot(df,vars=['bpm','nrgy','dnce','dB','live','val','dur','acous','spch','pop'],hue='top genre')
plt.show()
```
